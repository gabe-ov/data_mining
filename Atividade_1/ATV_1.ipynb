```python
search_response = youtube.search().list(
    q='eleição',
    type='video',
    regionCode='BR',
    publishedAfter='2022-10-02T00:00:00Z',
    publishedBefore='2022-10-30T23:59:59Z',
    part='snippet',
    maxResults=max_results,
    videoDuration='medium'
).execute()
```

```python
max_comment_count = 0
max_comment_video = None

for item in search_response.get('items', []):
    video_id = item['id']['videoId']

    video_info = youtube.videos().list(part='statistics', id=video_id).execute()

    if 'commentCount' in video_info['items'][0]['statistics']:
        comment_count = int(video_info['items'][0]['statistics']['commentCount'])

        if comment_count > max_comment_count:
            max_comment_count = comment_count
            max_comment_video = item
```

```python
if max_comment_video:
    print("Vídeo com mais comentários:")
    print("Título:", max_comment_video['snippet']['title'])
    print("ID do Vídeo:", max_comment_video['id']['videoId'])
    print("Número de Comentários:", max_comment_count)
else:
    print("Nenhum vídeo encontrado.")
```

```python
import requests
from bs4 import BeautifulSoup
```

```python
url = 'https://www.apontador.com.br/local/sp/campinas/faculdades_e_universidades/88Q2E372/unicamp.html'

response = requests.get(url)
```

```python
if response.status_code == 200:
    soup = BeautifulSoup(response.text, 'html.parser')

    review_elements = soup.find_all('div', class_='store-comments__item')

    for review in review_elements:
        author_name = review.find('span', class_='name').text

        publication_date = review.find('span', class_='rate-stars__amount').text

        stars = len(review.find_all('img', alt=''))

        review_text = review.find('div', class_='store-comments__text').p.text

        print(f'Autor: {author_name}')
        print(f'Data de Publicação: {publication_date}')
        print(f'Classificação em Estrelas: {stars}')
        print(f'Avaliação: {review_text}')
        print('---')

else:
    print(f'Erro na solicitação HTTP. Código de status: {response.status_code}')
```

```python
from selenium import webdriver
import time
```

```python
driver = webdriver.Firefox()

driver.get("https://www.uol.com.br/")

time.sleep(5)

last_height = driver.execute_script("return document.body.scrollHeight")
while True:
    driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
    time.sleep(2)
    new_height = driver.execute_script("return document.body.scrollHeight")
    if new_height == last_height:
        break
    last_height = new_height

# Fechar o navegador
driver.quit()
```
